### ner

Spacy und flair verhalten sich in diesem Beispiel gleich, also verwende einfach das, womit du dich am wohlsten fühlst.

```python
#imports
import requests
import io
import tika
from tika import parser
import string
import spacy

from flair.data import Sentence
from flair.models import SequenceTagger
```

In der nächsten Zelle musst du den Teil des Codes einkommentieren, der sich auf deine Situation bezieht:

```python
#wenn du die pdf nicht lokal speichern möchtest, die folgenden 4 Zeilen einkommentieren:
# url = "https://www.e-periodica.ch/cntmng?pid=grs-002%3A1984%3A76%3A%3A218"
# r = requests.get(url)
# f = io.BytesIO(r.content)
# parsed = parser.from_buffer(f)

#wenn du sie auf dem computer hast, dann diese Zeile einkommentieren:
parsed = parser.from_file('grs-002_1984_76__219_d.pdf')
```

Jetzt bereiten wir den Text vor

```python
contents = [x.strip() for x in parsed["content"].split("\n") if x != ""]
#remove the first page
article = contents[contents.index('https://www.e-periodica.ch/digbib/about3?lang=en')+1:]
article = " ".join(article)
```

#### Flair
Zunächst zeigen wir hier, wie das named entity tagging mit Flair funktioniert.

```python
# load tagger, this might take a while
tagger = SequenceTagger.load("flair/ner-german-large")

# predict on the article
sentence = Sentence(article)
tagger.predict(sentence)
```

```python
people = []
places = []
organisations = []

for entity in sentence.get_spans('ner'):
    if entity.tag == "PER": #people
        name = entity.text.translate(str.maketrans('', '', string.punctuation)) #remove possible ocr mistakes
        if len(name) >= 3: # names are usually not shorter
            people.append(entity.text)
    elif entity.tag == "LOC": #places
        place = entity.text.translate(str.maketrans('', '', string.punctuation)) #remove possible ocr mistakes
        if len(place) >= 3: # place are usually not shorter
            places.append(entity.text)
    elif entity.tag == "ORG": #only organisations
        org = entity.text.translate(str.maketrans('', '', string.punctuation)) #remove possible ocr mistakes
        if len(org) >= 3:
            organisations.append(entity.text)
```

```python
tagger_onto = SequenceTagger.load("flair/ner-english-ontonotes-large")
```

```python
#evaluate if you have ground truth (perfect transcription)
#here we take the eperiodica text as perfect transctiption
begin_idx = 0
end_idx = article.index("10")
# this is assuming that our gt is for the entire article, but our transcription is only for one page
# in my example this is what i need to search for to find the final index, yours might be different
gt = article[begin_idx:end_idx] 
```
